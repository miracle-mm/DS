```python
import pandas as pd
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999

df = pd.read_csv('MM Data - PartnersWithLatLongNewLabel.csv')
dd = pd.read_json('Partners_Table_with_email.json')

df['Volunteering'] =df['I am interested in'].str.contains('Volunteering')
df['Partnering'] =df['I am interested in'].str.contains('Partnering')
df['Media coverage'] =df['I am interested in'].str.contains('Media coverage')
df['Donating'] =df['I am interested in'].str.contains('Donating')
df['Something else'] =df['I am interested in'].str.contains('Something else')
df['Joining the MM team'] =df['I am interested in'].str.contains('Joining the MM team')
df['Partnering (for orgs/govt/etc)'] =df['I am interested in'].str.contains('Partnering (for orgs/govt/etc)')
df['Openings at MM'] =df['I am interested in'].str.contains('Openings at MM')

df = df.merge(dd, left_on='Email address', right_on='email', how='inner')

d1 = df.melt(id_vars=['id'], value_vars=['Volunteering'], value_name = 'Volunteering')
d1T = d1.loc[(d1['Volunteering']==True)]
d1T = d1T.rename(columns={'variable':'label'})
d1T = d1T.drop(columns='Volunteering')
d2 = df.melt(id_vars=['id'], value_vars=['Partnering'], value_name = 'Partnering')
d2T = d2.loc[(d2['Partnering']==True)]
d2T = d2T.rename(columns={'variable':'label'})
d2T = d2T.drop(columns='Partnering')
d3 = df.melt(id_vars=['id'], value_vars=['Media coverage'], value_name = 'Media coverage')
d3T = d3.loc[(d3['Media coverage']==True)]
d3T = d3T.rename(columns={'variable':'label'})
d3T = d3T.drop(columns='Media coverage')
d4 = df.melt(id_vars=['id'], value_vars=['Donating'], value_name = 'Donating')
d4T = d4.loc[(d4['Donating']==True)]
d4T = d4T.rename(columns={'variable':'label'})
d4T = d4T.drop(columns='Donating')
d5 = df.melt(id_vars=['id'], value_vars=['Something else'], value_name = 'Something else')
d5T = d5.loc[(d5['Something else']==True)]
d5T = d5T.rename(columns={'variable':'label'})
d5T = d5T.drop(columns='Something else')
d6 = df.melt(id_vars=['id'], value_vars=['Joining the MM team'], value_name = 'Joining the MM team')
d6T = d6.loc[(d6['Joining the MM team']==True)]
d6T = d6T.rename(columns={'variable':'label'})
d6T = d6T.drop(columns='Joining the MM team')
d7 = df.melt(id_vars=['id'], value_vars=['Partnering (for orgs/govt/etc)'], value_name = 'Partnering (for orgs/govt/etc)')
d7T = d7.loc[(d7['Partnering (for orgs/govt/etc)']==True)]
d7T = d7T.rename(columns={'variable':'label'})
d7T = d7T.drop(columns='Partnering (for orgs/govt/etc)')
d8 = df.melt(id_vars=['id'], value_vars=['Openings at MM'], value_name = 'Openings at MM')
d8T = d8.loc[(d8['Openings at MM']==True)]
d8T = d8T.rename(columns={'variable':'label'})
d8T = d8T.drop(columns='Openings at MM')
d8T.head(3)
```

```python
Services = pd.concat([d1T, d2T, d3T, d4T, d5T, d6T, d7T, d8T])
Services = Services.sort_values(by = ['id'])
Services = Services.rename(columns={'label':'ServiceName'})
```
